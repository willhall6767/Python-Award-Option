```python
def create_greeting(name, time_of_day):
    greeting = f"Good {time_of_day}, {name}!"
    return greeting


message1 = create_greeting("Will", "morning")
message2 = create_greeting("Josh", "afternoon")
message3 = create_greeting("Dexter", "evening")

print(message1)
print(message2)
print(message3)
```

```python
def calculate_rectangle_area(length, width):
    area=length*width
    return area

length = float(input("enter your length"))
width = float(input("enter your width"))

area = calculate_rectangle_area(length, width)
print(f"area of the rectangle is {area}")
```

```python
def format_name(first, middle, surname):
    formatted = f"{surname.upper()}, {first} {middle}"
    return formatted

first_name = input("Enter you first name:")
middle_name = input("Enter you second name")
surname = input("Enter you surname")

full_name = format_name(first_name, middle_name, surname)
print(f"formatted_name: {full_name}")


```
