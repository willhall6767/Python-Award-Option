```python
def scoreChallenge():
    '''Scorer validation challenge'''
    scores =[]
    while True:
        try:
            score = int(input('Enter a test score'))
            scores.append(score)
            response = input('Want to continue(y/n):')
            if response.lower() == "n":
                break
        except ValueError:
            print('Score needs to be a number')
    return scores
        
```

```python
scores = scoreChallenge()
for score in scores:
    print(score)
```
